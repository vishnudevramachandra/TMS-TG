# Steps below show the logic for plotting some example data

## 1. Setup the environment for plotting

```python
import numba as nb
import pandas as pd
from tms_tg import TMSTG
import figures.figure1.figure1a as fig1a

# Setting the number of threads for Numba
nb.set_num_threads(max(1, int(nb.config.NUMBA_NUM_THREADS // 1.25)))

# Define the selection parameters for plotting activity of neurons in MC(L5), SC(L5), thal(BZ) & thal(CZ) regions
epochs = (
        {'selectionParams': {'Epoch': {'Region': 'MC', 'Layer': 'L5'}}},
        {'selectionParams': {'Epoch': {'Region': 'SC', 'Layer': 'L5'}}},
        {'selectionParams': {'Epoch': {'Region': 'thal', 'Layer': None}, 'RecArea ': 'BZ'}},
        {'selectionParams': {'Epoch': {'Region': 'thal', 'Layer': None}, 'RecArea ': 'CZ'}},
    )

# File with a list of paths to data locations of all animals
animalList = r'G:\Vishnu\data\TMSTG\animalList.xlsx'

# Instantiate a TMSTG object using this list
tms = TMSTG.load(animalList)

# Load active neurons from previously computed statistical analysis
activeNeu = pd.read_pickle("./activeNeu_Cortex")
```

## 2. Plot raster and peristimulus activity trace for a randomly selected neuron from each group

```python
from importlib import reload

# Reload the module 'fig1a' to ensure any changes made to it are reflected
reload(fig1a)

# Plot using 'fig1a' module
fig1a.plot(tms,                   # TMSTG object
          activeNeu,              # Boolean array of active neurons
          kind='rasterAndPsfr',   # Type of plot to generate
          colParams=epochs,       # Selection parameters for different epochs
          xlim=(-50, 350),        # Limits for the x-axis
)
```
